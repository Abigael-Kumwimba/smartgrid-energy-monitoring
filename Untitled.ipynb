```python
df = pd.read_csv("household_power_consumption.csv")
df.head()
```

```python
df["datetime"] = pd.to_datetime(
    df["Date"] + " " + df["Time"],
    format="%d/%m/%Y %H:%M:%S"
)

df = df.drop(columns=["Date", "Time"])
df = df.set_index("datetime")

df.head()
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df = df.dropna()
```

```python
df_hourly = df.resample("h").mean()
df_hourly.head()
```

```python
df_hourly["target"] = df_hourly["Global_active_power"].shift(-1)
df_hourly = df_hourly.dropna()
```

```python
X = df_hourly[["Global_active_power", "Voltage", "Global_intensity"]]
y = df_hourly["target"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

mae, rmse
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,5))
plt.plot(y_test.values[:200], label="Valeurs réelles")
plt.plot(y_pred[:200], label="Prédictions")
plt.legend()
plt.title("Prédiction de la consommation électrique")
plt.xlabel("Temps")
plt.ylabel("Global active power")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
```

```python
models = {
    "Regression Lineaire": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42)
}
```

```python
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    results.append({
        "Modele": name,
        "MAE": mae,
        "RMSE": rmse
    })

results_df = pd.DataFrame(results)
results_df
```

```python
best_model = results_df.sort_values("RMSE").iloc[0]
best_model
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(results_df["Modele"], results_df["RMSE"])
plt.title("Comparaison des modèles selon RMSE")
plt.xlabel("Modèle")
plt.ylabel("RMSE")
plt.show()
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
linear_pred = linear_model.predict(X_test)

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

plt.figure(figsize=(14,5))
plt.plot(y_test.values[:200], label="Valeurs réelles", linewidth=2)
plt.plot(linear_pred[:200], label="Régression linéaire")
plt.plot(rf_pred[:200], label="Random Forest")
plt.legend()
plt.title("Comparaison des prédictions")
plt.xlabel("Temps")
plt.ylabel("Consommation prédite")
plt.show()
```
